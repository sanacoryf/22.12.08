```python

url = "https://raw.githubusercontent.com/nytimes/covid-19-data/master/us.csv"
download = requests.get(url).content
```

```python
print(download)
```

```python
import pandas as pd
import io
df = pd.read_csv(io.StringIO(download.decode()))
```

```python
df.head()
```

```python
df['cases_per_day'] = df['cases'].diff()
```

```python
df.head()
```

```python
df.loc[0, 'cases_per_day'] = 1
df['cases_per_day'] =  df['cases_per_day'].astype('int')
```

```python
df.head()
```

```python
import plotly.express as px
px.line(x = df['date'], y = df['cases_per_day'], labels = {'x' : 'Date','y':'Cases'}, title = "Cases per Day")
```

```python
import plotly.express as px
df = px.data.gapminder().query('year == 2007')
fig = px.choropleth(df, locations = 'iso_alpha',
                    color = "lifeExp",
                    hover_name = "country",
                    color_continuous_scale = px.colors.sequential.Plasma)

fig.show()
```
